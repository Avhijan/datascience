```python
import pandas as pd
import seaborn  as sns
import matplotlib.pyplot as plt
```

```python


# Load dataset
def load_repositary1(file_path):
  return pd.read_csv(file_path)


df = load_repositary1("./Health_heart_experimental.csv")
df.head(30)
```

This shows the data of the Health heart experiment data set in a list which includes age, sex, SysBp , DiaBp, weight in kgs, height in kgs, BMI and indication of health.

# LINE CHART

A line chart is a graphical representation of data where individual data points are connected by straight lines. It is typically used to show trends over time, comparing continuous data across categories, and visualizing how a variable changes in relation to another variable.

```python
sns.lineplot(x=df['age'],y=df['BMI'])
```


## Line-Chart BETWEEN BMI and Age

The above line-chart shows the graph between age and BMI. This graph shows the change in body mass index with respect to age.

---

# Bar plot

A bar graph is a visual representation used to compare different categories or groups. It uses rectangular bars, where the height or length of each bar is proportional to the value it represents. Bar graphs are helpful for comparing data across different categories.

```python
sns.countplot(data=df, x='indication')
```

## Bar plot between healthy and unhealthy people

This graph shows total healthy and unhealthy people.

---

## BOX-PLOT

A box plot is a graphical representation of a dataset's distribution, showing its median, quartiles, and outliers. 

```python
sns.boxplot(df['weightKg'])
```

This boxplot shows median and quartile however it has no outliers. Here Q2(median) is 70 kg. This shows 50 percent of the peopele are of weight between 55 to around 85 kgs.

```python
sns.boxplot(x='sex',y='age',hue='indication',data=df)
```

# Scatter-plot

A scatter plot is a type of graph that displays individual data points as dots on a two-dimensional plane. It is used to show the relationship or correlation between two continuous variables. 

```python
sns.scatterplot(x='heightCm', y='weightKg', hue='indication',data=df)
plt.show()
```

This shows that as height increases, weight tends to increase as well but only those with a balanced proportion of height and weight are considered to be in optimal health.
